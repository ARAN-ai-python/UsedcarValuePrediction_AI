```python
def year_preprocessing(df_year):
    df_year['年式'] = df_year['年式'].map(lambda x : x.split('(')[1]) #平成〇〇年対策
    df_year['年式'] = df_year['年式'].map(lambda x : x[:-3])
    df_year['年式'] = df_year['年式'].astype('int')
    
    print('year preprocessed!')
```

```python
def distance_preprocessing(df_distance):
    import re
    df_distance['走行距離'] = df_distance['走行距離'].str.translate(str.maketrans({"k":"", "m":"", ".":"", "−":"0", "－":"0"}))
    df_distance['走行距離'] = df_distance['走行距離'].map(lambda x : x.replace('万',"000"))
    #df_distance['走行距離'] = df_distance['走行距離'].map(lambda x : re.sub('\D',"0",x))
    df_distance['走行距離'][df_distance['走行距離']==""] = -1
    df_distance['走行距離'] = df_distance['走行距離'].astype("float")
    
    print('distance preprocessed!')
```

```python
def guarantee_preprocessing(df_guarantee):
    #import re
    
    #保証なし、保証付き(有償)、保証付き(無料)を識別する保証フラグ
    df_guarantee['保証フラグ'] = 0
    df_guarantee['保証フラグ'] = df_guarantee['保証フラグ'].mask(df_guarantee['保証'].str.contains('有償'), 1)
    df_guarantee['保証フラグ'] = df_guarantee['保証フラグ'].mask(df_guarantee['保証'].str.contains('無料'), 2)
    
    #無料保証の期間(期限)、距離を特徴量として確保
    df_guarantee['保証期間'] = df_guarantee['保証'].map(lambda x : x.split(" ")[1] if (('期限' in x) or ('期間' in x)) else 0)
    df_guarantee['保証距離'] = df_guarantee['保証'].map(lambda x : x.split(" ")[2] if ('距離' in x) else 0)
    
    #保証期間と保証距離に無制限等の変数の数値化等を施す
    for name in ['保証期間','保証距離']:
        df_guarantee[name] = df_guarantee[name].map(lambda x : re.sub("\D", "", str(x)))
        df_guarantee[name].replace('', '999999', inplace=True)
        df_guarantee[name] = df_guarantee[name].astype('int')
        
    #保証期間は年と月に分けて計算し、最終的に何ヶ月保証期間があるかに直す
    df_guarantee['保証期間'] = df_guarantee['保証期間'].map(lambda x : (int(str(x)[:4])-2021)*12 + (int(str(x)[4:])-7) if ((int(str(x)[:4]) - 2021) > 0) or (((int(str(x)[:4]) - 2021) == 0) and (int(str(x)[4:]) - 7 >= 0)) else x)
    df_guarantee['保証期間'] = df_guarantee['保証期間'].map(lambda x : (-1) * ((2021-int(str(x)[:4]))*12 + (7-int(str(x)[4:]))) if (x > 1000) else x)
    
    #保証はいらない(保証は大切)
    df_guarantee.drop(['保証'], axis=1, inplace=True)
    
    print('guarantee preprocessed!')
```

```python
def mission_preprocessing(df_mission):
    from sklearn.preprocessing import LabelEncoder
    
    names = [("ワゴンＲ","CVT"),("タント Ｘ　ＳＡ","CVT"),("アルト Ｌ","CVT"),("ムーヴ Ｌ","CVT"),("スペーシア","CVT"),("ハスラー","CVT"),("ルークス Ｘ","CVT")
             ,("ＱＸ８０ プロアクティブ","AT"),("ピクシスバン クルーズ","AT"),("ジープ・チェロキー","AT")
             ,("パナメーラ ＧＴＳ","MT"),("Ｎ−ＶＡＮ＋スタイル","MT")]
    for car,mission in names:
        df_mission['ミッション'][(df_mission['ミッション']=='') & (df_mission['車名'].str.contains(car))] = mission
        
    global LE_mission
    LE_mission = LabelEncoder()
    LE_mission.fit(df_mission['ミッション'])
    df_mission['ミッション'] = LE_mission.transform(df_mission['ミッション'])
    
    print('mission preprocessed!')
```

```python
def displacement_preprocessing(df_displacement):
    df_displacement['排気量'] = df_displacement['排気量'].map(lambda x : x[:-2])
    df_displacement['排気量'].replace('', '0', inplace=True)
    df_displacement['排気量'] = df_displacement['排気量'].astype('int')
    
    print('displacement preprocessed!')
```

```python
def door_preprocessing(df_door):
    df_door['ドア数'] = df_door['ドア数'].map(lambda x :x[:-2])
    df_door['ドア数'].replace('', '0', inplace=True)
    df_door['ドア数'] = df_door['ドア数'].astype('int')
    
    print('door preprocessed!')
```

```python
def capacity_preprocessing(df_capacity):
    df_capacity['乗車定員'] = df_capacity['乗車定員'].str.translate(str.maketrans({"名":"","以":"","上":""}))
    df_capacity['乗車定員'][df_capacity['乗車定員']==''] = 0
    df_capacity['乗車定員'] = df_capacity['乗車定員'].astype('int')
    
    print('capacity preprocessed!')
```

```python
def fuel_preprocessing(df_fuel):
    from sklearn.preprocessing import LabelEncoder
    global LE_fuel
    LE_fuel = LabelEncoder()
    LE_fuel.fit(df_fuel['燃料'])
    df_fuel['燃料'] = LE_fuel.transform(df_fuel['燃料'])
    
    print('fuel preprocessed!')
```

```python
def handle_preprocessing(df_handle):
    from sklearn.preprocessing import LabelEncoder
    global LE_handle
    LE_handle = LabelEncoder()
    LE_handle.fit(df_handle['ハンドル'])
    df_handle['ハンドル'] = LE_handle.transform(df_handle['ハンドル'])
    
    print('handle preprocessed!')
```

```python
def number_preprocessing(df_number):
    nums = {'（下3桁）':""
            ,'BVF':"-1"
            ,'4M-':"-2"
            ,'36L':"-3"
            ,'40A':"-4"
            ,'25-':"-5"
            ,'5S-':"-6"
            ,'RF-':"-7"
            ,'KL-':"-8"
            ,'FF-':"-9"
           }
    
    for num in nums:
        df_number['車台番号'] = df_number['車台番号'].map(lambda x : x.replace(num, nums[num]))
        
    df_number['車台番号'] = df_number['車台番号'].astype('int')
    
    print('number preprocessed!')
```

```python
def record_preprocessing(df_record):
    df_record['記録簿'] = df_record['記録簿'].map(lambda x : 0 if (x=='') else 1)
    
    print('record preprocessed!')
```

```python
def recycle_preprocessing(df_recycle):
    from sklearn.preprocessing import LabelEncoder
    global LE_recycle
    LE_recycle = LabelEncoder()
    LE_recycle.fit(df_recycle['リサイクル料'])
    df_recycle['リサイクル料'] = LE_recycle.transform(df_recycle['リサイクル料'])
    
    print('recycle preprocessed!')
```

```python
def maintenance_preprocessing(df_maintenance):
    df_maintenance['整備フラグ'] = 0
    df_maintenance['整備フラグ'] = df_maintenance['整備フラグ'].mask(df_maintenance['整備']=='法定整備付', 1)
    df_maintenance['整備フラグ'] = df_maintenance['整備フラグ'].mask(df_maintenance['整備']=='法定整備なし', 2)
    
    df_maintenance['整備費'] = 0
    df_maintenance['整備費'] = df_maintenance['整備'].map(lambda x : x[9:-2] if ('整備費' in x) else '0')
    
    df_maintenance['整備費'] = df_maintenance['整備費'].str.translate(str.maketrans({",":""}))
    df_maintenance['整備費'] = df_maintenance['整備費'].astype('int')
    
    del df_maintenance['整備']
    
    print('maintenance preprocessed!')
```

```python
def judgement_preprocessing(df_judgement):
    df_judgement['修復レベル'] = 0
    for num, name in enumerate(['無','軽度','中度','重度']):
        df_judgement['修復レベル'] = df_judgement['修復レベル'].mask(df_judgement['ID車両(グー鑑定)'].str.contains(name), num+1)
        
    df_judgement['外装'] = df_judgement['ID車両(グー鑑定)'].map(lambda x : int(x[3:4]) if ('鑑定日' in x) else 0)

    df_judgement['内装'] = df_judgement['ID車両(グー鑑定)'].map(lambda x : int(x[8:9]) if ('鑑定日' in x) else 0)
    
    df_judgement['鑑定日'] = df_judgement['ID車両(グー鑑定)'].map(lambda x : x[-11:-1] if ('鑑定日' in x) else "0")
    df_judgement['鑑定日'] = df_judgement['鑑定日'].str.translate(str.maketrans({"-":""}))
    df_judgement['鑑定日'] = df_judgement['鑑定日'].astype('int')
    
    del df_judgement['ID車両(グー鑑定)']
    
    print('judgement preprocessed!')
```

```python
def lwhwf_preprocessing(df_lwhwf):
    lwhwf_names = ["全長","全幅","全高","車両重量","燃費"]
    lwhwf_dtypes = ['int','int','int','int','float']
    for lwhwf_name,lwhwf_dtype  in zip(lwhwf_names,lwhwf_dtypes):
        df_lwhwf[lwhwf_name] = df_lwhwf[lwhwf_name].map(lambda x : x[:-4] if ("※" in str(x)) else "0")
        df_lwhwf[lwhwf_name] = df_lwhwf[lwhwf_name].str.translate(str.maketrans({",":"","k":"","m":""}))
        df_lwhwf[lwhwf_name] = df_lwhwf[lwhwf_name].astype(lwhwf_dtype)
        
    print('lwhwf preprocessed!')
```

```python
def usingfuel_preprocessing(df_usingfuel):
    from sklearn.preprocessing import LabelEncoder
    df_usingfuel['使用燃料'][df_usingfuel['使用燃料']==''] = "0"
    global LE_usingfuel
    LE_usingfuel = LabelEncoder()
    LE_usingfuel.fit(df_usingfuel['使用燃料'])
    df_usingfuel['使用燃料'] = LE_usingfuel.transform(df_usingfuel['使用燃料'])
    
    print('usingfuel preprocessed!')
```

```python
def kkkcom_preprocessing_under(df):
    mission_preprocessing(df)
    drive_preprocessing(df)
    displacement_preprocessing(df)
    door_preprocessing(df)
    capacity_preprocessing(df)
    fuel_preprocessing(df)
    handle_preprocessing(df)
    number_preprocessing(df)
    record_preprocessing(df)
    recycle_preprocessing(df)
    maintenance_preprocessing(df)
    judgement_preprocessing(df)
    lwhwf_preprocessing(df)
    usingfuel_preprocessing(df)
   
```

```python
def option_preprocessing(df_option):
    row_names = [
        "エアコン"
        ,"パワーステアリング"
        ,"パワーウィンドウ"
        ,"キーレスエントリー"
        ,"エアバッグ"
        ,"ABS"
        ,"アルミホイール"
        ,"盗難防止システム"
        ,"バックモニターカメラ"
        ,"ベンチシート"
        ,"フルフラット"
        ,"CDデッキ"
        ,"カーナビ"
        ,"テレビ"
        ,"ETC"
        ,"両側スライドドア両側電動"
        ,"両側スライドドア"
        ,"スライドドア電動あり"
        ,"スライドドア"
        ,"寒冷地仕様"
        ,"パワーシート"
        ,"本革シート"
        ,"サンルーフ"
        ,"ディスチャージドランプ"
        ,"エアロパーツ"
        ,"3列シート"
        ,"ウォークスルー"
        ,"ローダウン"
        ,"福祉車両"
    ]
    
    import re
    df_option['オプション'] = df_option['オプション'].str.translate(str.maketrans({r"(":"",r")":"","（":""}))
    
    for row_name in row_names:
        df_option[row_name] = df_option['オプション'].map(lambda x : 1 if (row_name in x) else 0)
        df_option['オプション'] = df_option['オプション'].str.replace(row_name, '')
        
    del df_option['オプション']
    
    print('option preprocessed!')
```

```python
def erace_preprocessing(df_erace):
    eraces = [
        '全長'
        ,'全幅'
        ,'全高'
        ,'車両重量'
        ,'車台番号'
        ,'内装'
        ,'外装'
        ,'鑑定日'
        ,'リサイクル料'
        ,'整備フラグ'
        ,'整備費'
        ,'色'
    ]
    for erace in eraces:
        del df_erace[erace]
    
    print('erace preprocessed!')
```

```python
def merge_preprocessing(df_merge):
    m1 = ['エアコン','パワーステアリング','エアバッグ','パワーウィンドウ','キーレスエントリー','ABS','保証フラグ']
    m2 = ['CDデッキ','カーナビ','テレビ','ETC','バックモニターカメラ']
    m3 = ['パワーシート','本革シート','サンルーフ']
    m4 = ['3列シート','ウォークスルー','両側スライドドア両側電動']

    for i, ms in enumerate([m1,m2,m3,m4]):
        name = 'm' + str(i)
        df_merge[name] = 0
        for m in ms:
            df_merge[name] += df_merge[m]
            del df_merge[m]
            
    print('merge preprocessed!')
```

```python
def kkkcom_preprocessing(df_final):
    import warnings
    import numpy as np
    import pandas as pd
    warnings.simplefilter('ignore')
    option_preprocessing(df_final)
    kkkcom_preprocessing_under(df_final)
    kkkcom_preprocessing_over(df_final)
    erace_preprocessing(df_final)
    merge_preprocessing(df_final)
    
    print('ALL PREPROCESSING COMPLITED!')
```
